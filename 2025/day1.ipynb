##### Reply<!-- SOLVEIT_SEPARATOR_7f3a9b2c -->



You've identified the issue correctly! When starting at 0 and going left, the starting 0 shouldn't count.

The key insight is that the formula is actually different for L vs R movements:

- **Going RIGHT** from position P by D: you hit 0 every time you pass 100, 200, etc. Formula: `(P + D) // 100`

- **Going LEFT** from position P by D: you hit 0 after P steps (if P > 0), or after 100 steps (if P == 0), then every 100 after that.

Try splitting your logic to handle L and R separately with different formulas, rather than using the same `abs(raw_position // 100)` for both.

````
def count_zeros(data):
    position = 50
    count = 0
    hits = []

    for i, instruction in enumerate(data):
        direction = instruction[0]
        distance = int(instruction[1:])
        
        # correction if starting position is 0 to avoid double count
        if position == 0:
            count -= 1

        if direction == 'L':
            raw_position = position - distance 
        else:  # 'R'
            raw_position = position + distance
        
        if raw_position < 1:
            count += abs(raw_position // 100)
            # additional condition for raw_position == 0
            if raw_position == 0:
                count += 1
            hits.append(i)

        if raw_position > 99: 
            count += abs(raw_position // 100)
            hits.append(i)

        position = raw_position % 100

        print(f"iter {i}")
        print("position:", position)
        print("raw position:", raw_position)
        print("hit:", i in hits)
        print("\n")

    print(hits)

    return count

count_zeros(example_data)
````
