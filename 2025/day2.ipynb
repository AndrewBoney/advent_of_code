Great question! The key insight: if a number is "doubled" like `6464` or `123123`, then its string representation has **even length** and the **first half equals the second half**.

So for any number `n`:
```python
s = str(n)
len(s) % 2 == 0 and s[:len(s)//2] == s[len(s)//2:]
```

Try this check on a few examples like `6464`, `123123`, and `123` to see if it works!

````
def pt1(data_ls):
    invalid_ids = []
    for rng in data_ls:
        l, r = rng.split("-")
        
        for i in range(int(l), int(r)+1):
            s = str(i) 
            if len(s) % 2 == 0:
                if s[:len(s)//2] == s[len(s)//2:]:
                    invalid_ids.append(int(s))

    sum_ = sum(invalid_ids)

    return invalid_ids, sum_
````
