Start with your original diagram. How many rolls of paper in total can be removed by the Elves and their forklifts?

Answer: 
 

````
from copy import deepcopy

def get_hits(grid):
    accessible = 0
    hits = []
    max_row = len(grid) - 1
    max_col = len(grid[0]) - 1
    
    for row in range(len(grid)):
        for col in range(len(grid[0])):
            if grid[row][col] != '@':
                continue
            
            roll_count = 0
            for dr, dc in movements:
                neighbour_row = row + dr
                neighbour_col = col + dc
                
                if (neighbour_row < 0) or (neighbour_row > max_row):
                    continue
                if (neighbour_col < 0) or (neighbour_col > max_col):
                    continue
                
                if grid[neighbour_row][neighbour_col] == "@":
                    roll_count += 1
            
            if roll_count < 4:
                accessible += 1
                hits.append((row, col))
    
    return accessible, hits

def remove_hits(grid, hits):
    grid_copy = deepcopy(grid)

    for row, col in hits:
        grid_copy[row][col] = "."

    return grid_copy

def pt2(grid):
    grid_copy = deepcopy(grid)

    total = 0
    count = float("inf")

    while count > 0:
        count, hits = get_hits(grid_copy)
        grid_copy = remove_hits(grid_copy, hits)
        total += count

    return total
````
