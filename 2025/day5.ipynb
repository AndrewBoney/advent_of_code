The key insight: instead of expanding ranges into individual numbers, work with the ranges themselves. You need to:

1. Merge overlapping ranges into non-overlapping ones
2. Sum up the lengths of the merged ranges

For example, ranges `3-5` and `10-14` don't overlap, so they contribute `3 + 5 = 8` IDs. But `10-14` and `12-18` overlap, so they merge into `10-18` which contributes `9` IDs.

Not quite! You don't need to check every range against every other range individually.

The trick is to **sort the ranges first** by their start positions. Then you can merge them in a single pass through the sorted list.

Here's the idea:
- Sort ranges by start value
- Keep track of the "current merged range"
- For each new range, check if it overlaps with or extends the current one
- If yes, extend the current range; if no, save it and start a new one

````
def pt2(ranges):
    sorted_ranges = sorted(ranges, key=lambda r: r[0])

    total = 0
    current_min, current_max = sorted_ranges[0]
    for i, (min_, max_) in enumerate(sorted_ranges):
        if min_ > current_max:
            total += (current_max - current_min) + 1
            current_min = min_
            current_max = max_

        if max_ > current_max:
            current_max = max_

    # final add
    total += (current_max - current_min) + 1

    return total
````
